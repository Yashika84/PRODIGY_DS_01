```python
# Step 1: Define year range
years = [str(y) for y in range(2001, 2023)]


male_ind = Data[(Data['Country Name'] == 'India') & (Data['Series Code'] == 'SP.POP.TOTL.MA.IN')]
female_ind = Data[(Data['Country Name'] == 'India') & (Data['Series Code'] == 'SP.POP.TOTL.FE.IN')]


male_values = male_ind[years].values.flatten() / 1e9
female_values = female_ind[years].values.flatten() / 1e9


# Step 2: Prepare data for plotting
df = pd.DataFrame({
    'Year': years * 2,
    'Population (Billions)': list(male_values) + list(female_values),
    'Gender': ['Male'] * len(years) + ['Female'] * len(years)
})

# Step 3: Create grouped bar plot
plt.figure(figsize=(16, 6))
sns.barplot(data=df, x='Year', y='Population (Billions)', hue='Gender', palette=['#1f77b4', '#e377c2'])

# Step 4: Customize and show plot
plt.title("India's Male and Female Population Over Time (2001–2022)", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Population (Billions)", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```
